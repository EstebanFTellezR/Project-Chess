```python
df_nombres=df[['Result','ECO','Opening','TimeControl','AN']]
df_nombres.to_csv('nombres.csv', index=False)
```

```python
df_nombres
```

```python
df_defensas = df_nombres[df_nombres['Opening'].str.contains('Defense', case=False, na=False)]
df_defensas.to_csv('defensas.csv', index=False)
```

```python
df_aperturas= df_nombres[~df_nombres['Opening'].str.contains('Defense', case=False, na=False)]
df_aperturas.to_csv('aperturas.csv', index=False)
```

```python
df_resultado=df_defensas[['Result']]
df_resultado
```

```python
# Crear una copia independiente del DataFrame para evitar advertencias
df_resultado = df_resultado.copy()

# Convertir la columna 'Result' a texto y limpiar espacios en blanco
df_resultado['Result'] = df_resultado['Result'].astype(str).str.strip()

# Verificar los valores Ãºnicos en la columna 'Result' despuÃ©s de la limpieza
print("Valores Ãºnicos en la columna 'Result' despuÃ©s de la limpieza:")
print(df_resultado['Result'].unique())

# Definir la funciÃ³n de mapeo
def map_result(result):
    result = result.strip()  # Limpiar espacios en blanco
    if result == '1-0':
        return 'victory'
    elif result == '0-1':
        return 'losses'
    elif result == '1/2-1/2':
        return 'draw'
    else:
        return 'unknown'

# Aplicar la funciÃ³n a la columna 'Result' usando .loc para evitar advertencias
df_resultado['Result'] = df_resultado['Result'].apply(map_result)

# Verificar algunos resultados despuÃ©s de aplicar la funciÃ³n
print("\nAlgunos valores despuÃ©s de aplicar la funciÃ³n de mapeo:")
print(df_resultado.head(10))

# Guardar el DataFrame actualizado en un nuevo archivo CSV
df_resultado.to_csv('resultados_actualizados.csv', index=False)

print('Columna Result actualizada y guardada en resultados_actualizados.csv.')
```
